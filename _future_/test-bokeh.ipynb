```python
from bokeh.io import push_notebook
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import time

source = ColumnDataSource(data={'x': [1, 2, 3, 4, 5], 'y': [6, 7, 2, 4, 5]})

p = figure(title="Dynamic Bokeh Plot")
p.line('x', 'y', source=source, line_width=2)
handle = show(p, notebook_handle=True)

# Dynamically update the data
for i in range(6, 10):
    source.stream({'x': [i], 'y': [i**0.5]})
    push_notebook(handle=handle)
    time.sleep(4)
```

```python
from vslab._future_.data import Data2D
from bokeh.plotting import figure, show
from bokeh.models import Slider, ColumnDataSource, CustomJS
from bokeh.layouts import row, column
import numpy as np
from bokeh.io import output_notebook

output_notebook()

# Load data
d = Data2D('/Users/vibhor/Downloads/vslabtest')
x = d.X
z = d.read_column(2)

# Initial setup
initial_row = 0
data_row = z[initial_row]

# Create data sources
image_source = ColumnDataSource(data={"image": [z]})
line_source = ColumnDataSource(data={"x": x, "y": data_row})

# Create an image figure
image_figure = figure(title="Image Plot", width=300, height=300)
image_figure.image(image='image', x=0, y=0, dw=10, dh=10, palette="Spectral11", source=image_source)

# Create a line figure
line_figure = figure(title="Line Plot", width=300, height=300)
line_figure.line('x', 'y', source=line_source, line_width=2)

# Slider with a JavaScript callback
slider = Slider(start=0, end=z.shape[0]-1, value=0, step=1, title="Row Selector")
callback = CustomJS(args=dict(source=line_source, x=x, z=z), code="""
    const row = cb_obj.value;
    source.data = { x: x, y: z[row] };
    source.change.emit();
""")
slider.js_on_change('value', callback)

# Combine layouts
layout = column(slider, row(image_figure, line_figure))

# Show inline in Jupyter
show(layout)
```

```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, Slider

output_notebook()

# Prepare the data
x = list(range(1, 101))
y = [i**2 for i in x]
source = ColumnDataSource(data=dict(x=x, y=y))

# Create the figure
p = figure(title="Interactive Quadratic Function", x_axis_label='x', y_axis_label='y')
p.line('x', 'y', source=source, line_width=3)

# Create the slider
slider = Slider(start=0, end=10, value=1, step=0.1, title="Coefficient")

# Callback function to update the plot
def update_data(attrname, old, new):
    a = slider.value
    y = [a * i**2 for i in x]
    source.data = dict(x=x, y=y)

# Attach the callback to the slider
slider.on_change('value', update_data)

# Display the plot and slider
show(p)
show(slider)
```
